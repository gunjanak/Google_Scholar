```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
from time import sleep
import time
from bs4 import BeautifulSoup
from urllib.request import urlopen
from bs4 import BeautifulSoup
import pandas as pd
```

```python
query = input("Enter what you want to search")
```

```python
query= query.replace(" ", "+")
query
```

```python
df = pd.DataFrame(columns=['UID','Title',"Article_Link",'Authors', 'Abstract', 'Publication', 'Num_citation', 'Cited_by_url', 'Published_Year'])
```

```python
df.head()
```

```python
from scholarLib import get_title,get_id,get_abstract,get_authors_publication,get_citations_no
```

```python
browser = webdriver.Firefox()

for i in range(0,1000,10):


    try:
        # path = f"https://scholar.google.com/scholar?start={i}&q=dnn&hl=en&as_sdt=0,5"
        path = f"https://scholar.google.com/scholar?start={i}&q={query}&hl=en&as_sdt=0,5"
        browser.get(path)
        html_content = browser.page_source
        if(i==0):
            time.sleep(65)
        else:
            time.sleep(2)

        titles,links = get_title(html_content)
        unique_ids = get_id(html_content)
        authors,publication,years = get_authors_publication(html_content)
        abstracts = get_abstract(html_content)
        citations,links_to_citations = get_citations_no(html_content)
        data = {
            'UID':unique_ids,
            'Title': titles,
            'Article_Link': links,
            'Authors': authors,
            'Publication': publication,
            'Published_Year': years,
            'Abstract': abstracts,
            'Num_citation': citations,
            'Cited_by_url': links_to_citations
        }
        # # Create DataFrame from dictionary
        # df = pd.DataFrame(data)

        # Convert additional data to DataFrame
        df_more = pd.DataFrame(data)
        

        # Append the new data to the existing DataFrame
        df = df._append(df_more, ignore_index=True)
        
    except Exception as e:  # Generic except block
        print(f"Error: {type(e).__name__} occurred!")
    
time.sleep(5)
browser.quit()
df.head()
path = query+".csv"
# Save the DataFrame to a CSV file
df.to_csv(path, index=False)

print("DataFrame saved as CSV file.")
```
